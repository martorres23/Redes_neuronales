```python

# Entrena una red MLP que aprenda la funci√≥n dada por la expresi√≥n:
#ùëì(ùëã) =1/2 ùë†ùëñùëõ(ùë•) + 1/2 ùë†ùëñùëõ (2ùë•), 
#d√≥nde ‚àíùúã < ùë• < ùúã. Para validar el entrenamiento, 
#verifique la generalizaci√≥n de la red de entrada.
```

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Generamos los datos de entrenamiento
x = (np.arange(-np.pi, np.pi, 0.2)).reshape(32,1)
print(x)
```

```python
y = (0.5*np.sin(x) + 0.5 * np.sin(2*x)).reshape(32,)
print(y)
```

```python
import matplotlib.pyplot as plt
# Se despliega la grafica X,y
plt.plot(x,y)
```

```python
# Seleccionamos un conjunto de entrenamiento y uno de test. 
X_train, X_test, y_train, y_test =  train_test_split(x,y,random_state=3, test_size = 0.3)
```

```python
from sklearn.neural_network import MLPRegressor

# Se crea un perceptron multicapa para regresi√≥n. 
MLPreg = MLPRegressor(solver = 'lbfgs', max_iter=10000, alpha= 1e-9, hidden_layer_sizes = (50,50,50))
```

```python
# se ajusta o entrena el MLP con los datos E/S de entrenamiento
MLPreg.fit(X_train,y_train)
```

```python
# Se calcula  e imprime el coeficiente de determinaci√≥n de la predicci√≥n.
score = MLPreg.score(X_test,y_test)
print ("La precisi√≥n media es ",score)
```

```python
# Creamos un nuevo conjunto de prueba, X_T
X_T = np.arange(-np.pi,np.pi,0.2).reshape(32,1)
y_T = (0.5*np.sin(x) + 0.5*np.sin(2*x)).reshape(32,)
```

```python
# Se aproxima el seno de X_ con el perceptron previamente entrenados
y_S = MLPreg.predict(X_T)
# Mostramos la gr√°fica de la aproximaci√≥n
plt.plot(X_T,y_S)
```

## An√°lisis de resultados:

La gr√°fica muestra que la aproximaci√≥n del seno utilizando el perceptr√≥n multicapa
se ajusta bastante bien a la funci√≥n original, y la precisi√≥n media obtenida es cercana a 1,
los resultados obtenidos indican que la red MLP ha aprendido correctamente
la funci√≥n ùëì(ùëã) y es capaz de generalizar bien a datos nuevos.

```python
# Entrene un MLP con el algoritmo backpropagation, para que aprenda
# la funci√≥n de dos variables mostrada en la gr√°fica. 
```

```python
#Se crea un arreglo de datos desde -3 hasta 3, tomando valores cada 0.2
x = np.arange(-3,3,0.2)
y = x;
# Se contruye una malla para despliegue en 3D
X,Y = np.meshgrid(x, y)

Z = (3*(1-X)**2*np.exp(-(X**2)-(Y+1)**2)
    -10*(X/5-X**3-Y**5)*np.exp(-X**2-Y**2)
     -1/3*np.exp(-(X+1)**2-Y**2))

#Despliegue de la imagen 3D
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
# Se transforma la matriz de cada malla en vector
X_grid = X.reshape(900,1)
y_grid = Y.reshape(900,1)
z_grid = Z.reshape(900,)
```

```python
# Se conforma una matriz de entrada con X_grid y y_grid, cada uno aporta
# una caracteristica en cada dato de entrada, es de dimension 900x2, es
# decir 900 datos, de 2 caracter√≠sticas cada uno.

P_in = P_in = np.concatenate((X_grid,y_grid),axis=1)
```

```python
# Se divide el conjunto de datos en entrenamiento y prueba
P_train, P_test, z_train, z_test = train_test_split(P_in, z_grid, test_size=0.3)
```

```python
# Se crea la red neuronal de regresi√≥n
net = MLPRegressor(solver = 'lbfgs', max_iter=20000, alpha = 1e-3, hidden_layer_sizes = (100,100,100))
```

```python
# Se entrena la red con el conjunto de entrenamiento
net.fit(P_train, z_train)
```

```python
# Se eval√∫a el desempe√±o de la red con el conjunto de prueba
z_pred = net.predict(P_test)
```

```python
# Se calcula el error de generalizaci√≥n
mse = np.mean((z_test - z_pred)**2)
mse
```

```python
# Se simula la funci√≥n mediante la red, para ver si puede aproximarla adecuadamente
P_grid = net.predict(P_in)
```

```python
# La salida es un vector unidimensional, debe ponerse en forma de matriz para
# poder desplegar la gr√°fica 3D.
P = P_grid.reshape(30,30)
```

```python
# Desplegamos la funci√≥n que aprendi√≥ la red
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, P)
plt.show()
```

```python
# 5.Entrenar un perceptr√≥n multicapa para clasificaci√≥n (MLPClassifier) y determinar
# mediante validaci√≥n cruzada, la exactitud en la clasificaci√≥n para la base de datos 
# de C√°ncer contenida en el archivo Cancer.zip. Verificar con qu√© modificaciones sobre
# el dise√±o de la red se obtienen mejores resultados. Comparar con los obtenidos mediante
# los clasificadores anteriores. Probar otros clasificadores implementados en la librer√≠a 
#scikit Learn. (Ver ejemplo en clase).
```

```python
#      Attribute                     Domain
#   -- -----------------------------------------
#   1. Sample code number            id number
#   2. Clump Thickness               1 - 10
#   3. Uniformity of Cell Size       1 - 10
#   4. Uniformity of Cell Shape      1 - 10
#   5. Marginal Adhesion             1 - 10
#   6. Single Epithelial Cell Size   1 - 10
#   7. Bare Nuclei                   1 - 10
#   8. Bland Chromatin               1 - 10
#   9. Normal Nucleoli               1 - 10
#  10. Mitoses                       1 - 10
#  11. Class:                        (2 for benign, 4 for malignant)


import pandas as pd
column_names=['Sample code number','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape',
             'Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin',
             'Normal Nucleoli','Mitoses','Class']
df = pd.read_csv('Cancer/breast-cancer-wisconsin.data', header=None, na_values='?', names=column_names)
# Reemplazar los valores faltantes por la media de la columna correspondiente
df.fillna(df.mean(), inplace=True)
df.head(10)
```

```python
# Creamos conjunto de prueba y entrenamiento.
x=df[['Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape',
             'Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin',
             'Normal Nucleoli','Mitoses']]
y=df['Class']
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(x,y, random_state=5, test_size = 0.3)
```

```python
# Creamos un objeto perceptron multicapa para clasificaci√≥n
MLP_cla = MLPClassifier(activation='relu', max_iter=80000, hidden_layer_sizes=(200,200,200))
```

```python
# Entrenamos el modelos
MLP_cla.fit(X_entrenamiento,y_entrenamiento.values.ravel())
```

```python
# Calculamos las predicciones con el conjunto de prueba
prediccion = MLP_cla.predict(X_prueba)
```

```python
# Calculamos la exactitud en la predicci√≥n

from sklearn.metrics import accuracy_score

accuracy_score(y_prueba, prediccion)
```

```python
# Calculamos la exactitud de la predicci√≥n con el m√©todo score del clasificador
MLP_cla.score(X_prueba, y_prueba)
```

```python
# Calculamos la matriz de confusi√≥n
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_prueba, prediccion)

skcm = pd.DataFrame(matriz, columns=['predice-Benigno','predice-Maligno'])
skcm['actual'] = ['Benigno','Maligno']
skcm = skcm.set_index('actual')
print(skcm)
```

```python
# Verificamos las m√©tricas de clasificacion

from sklearn.metrics import classification_report

print("\nReporte de la Clasificaci√≥n:")
print(classification_report(y_prueba, prediccion, target_names=['Benigno', 'Maligno']))
```

```python
#Validaci√≥n cruzada
from sklearn.model_selection import cross_val_score
scores = cross_val_score(MLP_cla, x, y, cv=5)
scores
print('Exactitud en la clasificaci√≥n: %0.2f (+/- %0.2f)' % (scores.mean(), scores.std() * 2))

#puntajes obtenidos en cada partici√≥n
print("Puntajes de validaci√≥n cruzada:", scores)

#promedio y desviaci√≥n est√°ndar de los puntajes
print("Promedio de puntajes:", scores.mean())
print("Desviaci√≥n est√°ndar de puntajes:", scores.std())
```

## An√°lisis del punto 5

Si aumentamos el tama√±o de capas ocultas y n√∫mero de iteraciones, adem√°s si utilizamos funciones de activaci√≥n como 'relu' √≥ 'Logistic' se obtiene un mejor rendimiento.

```python
# importamos los modelos a aplicar
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Creamos una lista de modelos
modelos = []
modelos.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
modelos.append(('LDA', LinearDiscriminantAnalysis()))
modelos.append(('KNN', KNeighborsClassifier()))
modelos.append(('CART', DecisionTreeClassifier()))
modelos.append(('NB', GaussianNB()))
modelos.append(('SVM', SVC(gamma='auto')))
modelos.append(('MPL', MLPClassifier(activation='tanh', max_iter=5000, hidden_layer_sizes=(10,10,10))))
```

```python
# Se procede a evaluar cada modelo
from sklearn.model_selection import cross_val_score
resultados = []
nombres = []
for nombre, modelo in modelos:
    cv_res = cross_val_score(modelo, X_entrenamiento, y_entrenamiento.values.ravel(), cv=5)
    resultados.append(cv_res)
    nombres.append(nombre)
    print("Clasificador = {}, Promedio = {}, Desviacion = {} ".format(nombre, cv_res.mean(), cv_res.std()))
```
